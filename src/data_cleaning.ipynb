### 4. Edicion de formato 

El principal objetivo de este punto es homojeinizar el formato de las columnas para poder acceder a ellas sin problemas 

```python
# 4.1 analisis de columnas para ver "como estan escrito "
df.columns
```

Se aprecia eterogeinidad en las etiquetas, series con espacios en blanco, mayusculas y minusculas entrmezcladas.
Esto podria iducir a errores al realizar consultas.

```python
# 4.2 Edicion de nombres de columnas. Pasamaos la primera letra de todos los nombres a mayuscula
df.columns = df.columns.str.title()
# comprobacion 
df.columns
```

```python
# 4.3 Eleminamos los espacios en blanco 
df.columns = df.columns.str.strip()
#comprobacion 
df.columns
```

```python
# 4.4 edicion de formato, remplazar caracteres
df.columns = df.columns.str.replace(" ", "_")
#comprobacion 
df.columns
```

```python
# 4.5 edicion de formato, remplazar y homogeinizar registros
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
df = df.apply(lambda x: x.str.title() if x.dtype == "object" else x)
#comprobacion 
df["Fatal_(Y/N)"].value_counts()


```
